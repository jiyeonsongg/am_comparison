```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/PSEN1_Sun.csv")

print(df.columns.tolist())
```

```python
mutation_col = "Mutation"
ratio_col = "Ab42/Ab40 ratio"
sd_col = "SD of ratio"

df = df.sort_values(by=ratio_col, ascending=False)
```

```python
# Define the list of mutations you want to include
selected_mutations = ["A246E", "I229F", "V272A"]

# Filter your dataset
filtered_df = df[df["Mutation"].isin(selected_mutations)]

# Check result
print(filtered_df)
```

```python
plt.figure(figsize=(10,6))
plt.bar(
    filtered_df["Mutation"],
    filtered_df["Ab42/Ab40 ratio"],
    yerr=filtered_df["SD of ratio"],
    capsize=3,
    color="#76a5af",
    alpha=0.8
)

# Labels and styling
plt.title("Ratio of Aβ42 over Aβ40 by Mutation", fontsize=14)
plt.xlabel("Mutation", fontsize=12)
plt.ylabel("Aβ42 / Aβ40 Ratio", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.6)
```

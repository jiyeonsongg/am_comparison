# **Statistical Analyses**

```python
import re
import pandas as pd

# --- FILE PATHS (edit if your filenames differ) ---
mut_fp = "/content/Supplemental Data - PSEN1 - Sun et al (1) (1).csv"  # has 'Mutation'
rsa_fp = "/content/PSEN1_AF3.csv"                                 # has 'ResNum','AA','RSA'

# --- LOAD ---
mut = pd.read_csv(mut_fp)
rsa = pd.read_csv(rsa_fp)

# keep just what's needed from RSA table
rsa = rsa.rename(columns=str.strip)
rsa = rsa[["ResNum", "AA", "RSA"]].copy()
rsa["ResNum"] = pd.to_numeric(rsa["ResNum"], errors="coerce")
rsa["AA"] = rsa["AA"].astype(str).str.strip().str.upper()

# --- PARSE Mutation like "R35Q" ---
def parse_mut(s):
    s = str(s).strip().upper()
    m = re.match(r'^[A-Z](\d+)[A-Z]$', s)
    if not m:
        # more flexible: capture WT, Pos, Mut separately
        m2 = re.match(r'^\s*([A-Z])\s*(\d+)\s*([A-Z])\s*$', s)
        if not m2:
            return pd.Series([None, None, None])
        return pd.Series([m2.group(1), int(m2.group(2)), m2.group(3)])
    return pd.Series([s[0], int(m.group(1)), s[-1]])

if "Mutation" not in mut.columns:
    raise ValueError("Could not find 'Mutation' column in the mutations file.")

parsed = mut["Mutation"].apply(parse_mut)
parsed.columns = ["WT", "Pos", "Mut"]
mut2 = pd.concat([mut, parsed], axis=1)

# --- MERGE to bring in RSA by position ---
merged = mut2.merge(rsa, left_on="Pos", right_on="ResNum", how="left")

# --- CHECKS ---
merged["WT"] = merged["WT"].astype(str).str.upper()
merged["Mut"] = merged["Mut"].astype(str).str.upper()
merged["WT_matches_RSA_AA"] = (merged["WT"] == merged["AA"])

# --- QUICK PREVIEW ---
preview_cols = ["Mutation", "WT", "Pos", "Mut", "AA", "RSA", "WT_matches_RSA_AA"]
print("Preview (first 10 rows):")
display(merged[preview_cols].head(10))

print("\nCounts:")
print("  Total rows:", len(merged))
print("  With RSA matched by position:", merged["RSA"].notna().sum())
print("  WT matches RSA.AA:", merged["WT_matches_RSA_AA"].sum())

print("\nExamples where WT != RSA.AA (first 10):")
display(merged.loc[merged["RSA"].notna() & ~merged["WT_matches_RSA_AA"], preview_cols].head(10))
```

```python
import pandas as pd
import numpy as np

# Load your processed file
df = pd.read_csv("/content/PSEN2_Pillai_RSA_resolved.csv")

# Keep only relevant columns
df = df[["AA", "RSA"]].dropna()
df["AA"] = df["AA"].astype(str).str.strip().str.upper()

# Compute group-level stats
rsa_stats = (
    df.groupby("AA")["RSA"]
      .agg(mean_RSA="mean", sem_RSA=lambda x: np.std(x, ddof=1)/np.sqrt(len(x)), n="count")
      .reset_index()
      .sort_values("AA")
)

print(" Descriptive RSA statistics by amino acid:")
print(rsa_stats.to_string(index=False))

# Optionally save
rsa_stats.to_csv("/content/PSEN2_Pillai_RSA_stats.csv", index=False)
print("\n Saved summary  /content/PSEN2_Pillai_RSA_stats.csv")
```
